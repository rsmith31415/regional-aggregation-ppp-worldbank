```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')


fig, [ax,ax0,ax1] = plt.subplots(3,1, figsize=(15,30))

df = pd.read_csv('RegionalTable_1.9.csv')
df[['requestYear', 'regionTitle', 'hc']].groupby(['requestYear', 'regionTitle']).sum()['hc'].unstack().plot(ax=ax)
ax.legend(loc="lower left")
ax.set_title('$1.9/Day')
ax.xaxis.label.set_visible(False)

df1 = pd.read_csv('RegionalTable_3.2.csv')
df1[['requestYear', 'regionTitle', 'hc']].groupby(['requestYear', 'regionTitle']).sum()['hc'].unstack().plot(ax=ax0)
ax0.legend(loc="lower left")
ax0.set_title('$3.2/Day')
ax0.xaxis.label.set_visible(False)

df2 = pd.read_csv('RegionalTable_5.5.csv')
df2[['requestYear', 'regionTitle', 'hc']].groupby(['requestYear', 'regionTitle']).sum()['hc'].unstack().plot(ax=ax1)
ax1.legend(loc="lower left")
ax1.set_title('$5.5/Day')
ax1.xaxis.label.set_visible(False)
```
